```python
vectA = np.array([8,4,2])
vectB = np.array([3,-1,-4])
vectC = np.array([7,0,-5])

R = np.arange(-10,10,1)

a1, a2, a3 = np.meshgrid(R,R,R)

vectR = vectA + vectB + vectC
spanRx = a1*vectA[0] + a2*vectB[0] +  a3*vectC[0]
spanRy = a1*vectA[1] + a2*vectB[1] +  a3*vectC[1]
##plt.scatter(R*vectA[0],R*vectA[1])
##plt.scatter(R*vectB[0],R*vectB[1])
plt.scatter(spanRx,spanRy, s=5, alpha=0.75)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```
